```python
import os
import sys
import mne

project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))
sys.path.append(project_root)

from src.preprocessing.data_pipeline import DataPipeline
from src.preprocessing.feature_extractor import FeatureExtractor

# Test data pipeline
pipeline = DataPipeline("../data/raw")
processed_data, labels = pipeline.process_dataset()

# Debug prints
print(f"Type of processed_data: {type(processed_data)}")
if processed_data:
    print(f"Type of first element: {type(processed_data[0])}")
    print(f"Length of processed_data: {len(processed_data)}")

# Test feature extraction
extractor = FeatureExtractor()
if processed_data and isinstance(processed_data[0], mne.io.edf.edf.RawEDF):  # Changed this line
    features = extractor.extract_features(processed_data[0])
    print("Features extracted successfully")
else:
    print("No valid data found or incorrect data type")
    
# Print feature structure
if features:
    print("\nExtracted Features:")
    for channel, channel_features in features.items():
        print(f"\nChannel: {channel}")
        print("Power Bands:")
        for band, power in channel_features['power_bands'].items():
            print(f"  {band}: {power:.2f}")
        print("Statistical Features:")
        for stat, value in channel_features['statistical'].items():
            print(f"  {stat}: {value:.2f}")
```
